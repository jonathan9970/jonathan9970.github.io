```python
import string
text="The University of Missouri–St. Louis (UMSL) is a large public research university in St. Louis County known for combining academic opportunity with strong community engagement. Founded in 1963 as part of the University of Missouri System, it serves a diverse student body and offers a wide range of undergraduate, graduate, and doctoral programs, including Missouri’s only professional Doctor of Optometry degree. UMSL emphasizes accessibility, research, and regional impact, with more than 105,000 alumni—most living and working in the St. Louis area—and a campus that blends urban energy with a commitment to student success. "

print(string.punctuation)

for p in string.punctuation:
  text=text.replace(p,'')

text=text.lower()

words=text.split()
print(text)

freq={}
for word in words:
  if word in freq:
    freq[word]+=1
  else:
    freq[word]=1

print(freq)

print(freq.items())

freq_sorted=sorted(freq.items(), key=lambda x:x[1], reverse=True)
print(freq_sorted)

top5 = freq_sorted[:5]
print("top 5 -->",top5)
```

```python
student_scores = {
"Alice": [88,92,79],
"Bob": [90,85,95],
"Charie": [70,80,85]
}

def letter_grade(score):
  if score >=90:
    return "A"
  elif score >=80:
    return "B"
  elif score >=70:
    return "C"
  elif score >=60:
    return "D"
  else:
    return "F"

result = {}

for student, scores in student_scores.items():
  average = sum(scores)/len(scores)
  result[student] = (average, letter_grade(average))

print(result)

sorted_students = sorted(result.items(), key=lambda x: x[1], reverse=True)
print("tope student is", sorted_students[0])
```

```python
inventory = {
"laptop": {"price": 1200, "quantity": 4},
"mouse": {"price": 25, "quantity": 20},
"keyboard": {"price": 75, "quantity": 10}
}

total_value = 0
values = []

print(inventory.items())

for item, details in inventory.items():
  price = details["price"]
  quantity = details["quantity"]
  value = price * quantity
  values.append((item, value))
  total_value += value
print(values)
print("total value of this inventory is",total_value)

#find most valuable item

most_valuable_item = max(values, key=lambda x: x[1])
print("most valuable item is", most_valuable_item)

#sort list of products by values
sorted_products = sorted(values, key=lambda x: x[1], reverse=True)
print("sorted products", sorted_products)
```
