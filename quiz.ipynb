```python
x = int(input('Score:'))

if x>=93:
    grade ='A'
elif x>=90:
    grade ='A-'
elif x>=87:
    grade ='B+'
elif x>=83:
    grade ='B'
elif x>=80:
    grade ='B-'
elif x>=77:
    grade ='C+'
elif x>=73:
    grade ='C'
elif x>=70:
    grade ='C-'
elif x>=67:
    grade ='D+'
elif x>=63:
    grade ='D'
elif x>=60:
    grade ='D-'
else:
    grade ='F'
print(grade)
```
